## Import libraries:

```python
# import libraries
import pandas as pd
import seaborn as sns 
import matplotlib as plt

# X and Y - Train & Test
from sklearn.model_selection import train_test_split

# Pipeline - Vectorization (Text to Number) - Model Selection - Train Model
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Evaluation - Prediction
from sklearn.metrics import classification_report
```

# Load Dataset:

```python
df = pd.read_csv('Fake_Real_News_Data.csv')
df.drop(columns="Unnamed: 0",inplace=True)
# df.info()
df.head()
# df.shape
```

## Check Missing Values:

```python
# df.isnull().sum()
df.dropna(axis=1)
```

## Check Duplicate Data:

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

## Select X and Y / Train Test Split:

```python
df['Text'] = df['title'] + df['text']

X = df['text']
y = df['label']
xtrain, xtest, ytrain, ytest = train_test_split(X, y, train_size=0.8, random_state=42)
```

## Check Balanced / Imbalanced:

```python
y.value_counts(normalize=True) * 100
```

## Model Selection / Train Model:

```python
main_pipe = Pipeline (                                          # Text Vectorization
   steps= [
            ('tfidf',TfidfVectorizer()),
            ('model',LogisticRegression(max_iter=1000))
    ]
)

main_pipe.fit(xtrain,ytrain)                                    # Train Model
```

## Evaluation / Prediction:

```python
print("Train: ")
ytrain_pred = main_pipe.predict(xtrain)
print(classification_report(ytrain,ytrain_pred))

print("test: ")
ytest_pred = main_pipe.predict(xtest)
print(classification_report(ytest,ytest_pred))
```
